- **Calculate ride duration**

```python
df['duration_minutes'] = ((df['ended_at'] - df['started_at'])/pd.Timedelta(minutes=1)).astype('int32')
```

- **Start and End Location**

```python
# Combining lat and lng for coordinates

df['start_loc'] = list(zip(df['start_lat'], df['start_lng']))
df['end_loc'] = list(zip(df['end_lat'], df['end_lng']))
```

-  **Distance between start and end stations**

```python
# Function to calculate the distance between stations

def distance_formula(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in kilometers
    lat1, lon1, lat2, lon2 = radians(lat1), radians(lon1), radians(lat2), radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Create column
df['distance_km'] = np.vectorize(distance_formula)(
    df['start_lat'],
    df['start_lng'],
    df['end_lat'],
    df['end_lng'])
```

- **Month, Day and start Hour**

```python
month_order = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 
         7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

df['month'] = df['started_at'].dt.month.map(month_order)

df['day'] = df['started_at'].dt.day_name()

df['hour'] = df['started_at'].dt.hour
```

```python
# Dataframes of station names and station locations

start_station = df[['start_station_name', 'start_loc', 'member_casual']].copy()

end_station = df[['end_station_name', 'end_loc', 'member_casual']].copy()

# Renaming columns

start_station.rename(columns={'start_station_name': 'station_name', 'start_loc': 'station_loc'}, inplace=True)
end_station.rename(columns={'end_station_name': 'station_name', 'end_loc': 'station_loc'}, inplace=True)

# Exclude rows where start or end station is defined 'not_at_station'
start_station = start_station[start_station['station_name'] != 'not_at_station']
end_station = end_station[end_station['station_name'] != 'not_at_station']

# Combine the entries of start and end stations into a single DataFrame called all_station

all_station = pd.concat([start_station, end_station], ignore_index=True).reset_index(drop=True)

all_station
```

```python
# Obtain the most common location and occurrence count per station for casual users

station_casual = all_station.query('member_casual == "casual"')\
                           .groupby('station_name')['station_loc']\
                           .agg([('station_loc', 'first'), ('occurrence', 'size')])\
                           .reset_index()\
                           .sort_values(by='occurrence', ascending=False)\
                           .reset_index(drop=True)

station_casual
```

```python
# Obtain the most common location and occurrence count per station for members

station_member = all_station.query('member_casual == "member"')\
                            .groupby('station_name')['station_loc']\
                            .agg([('station_loc', 'first'), ('occurrence', 'size')])\
                            .reset_index()\
                            .sort_values(by='occurrence', ascending=False)\
                            .reset_index(drop=True)

station_member
```

<a id="ride-count"></a>
</div>

- In this section I will create visualizations to showcase the number of rides taken by Casual and Member users throughout the year, broken down by month, day, and hour.

```python
agg = pd.DataFrame(df.groupby('member_casual')['duration_minutes']
                   .agg(['count', 'mean'])
                   .reset_index()
                   .rename(columns={'member_casual': 'user_type',
                                    'count': 'ride_count',
                                    'mean': 'mean_duration'}))

# Ride Count Pie Chart
fig, ax1 = plt.subplots(figsize=(6, 6))
ax1.set_title('Ride Count (2022)', fontsize=16, fontweight='bold')
ride_counts = agg['ride_count']
wedges, texts, autotexts = ax1.pie(ride_counts, labels=agg['user_type'], autopct='%1.0f%%',
                                   startangle=140, wedgeprops={'edgecolor': 'black'})
ax1.axis('equal')

legend_labels = [f"{agg['user_type'][i]}: {ride_counts[i]}" for i in range(len(agg))]
ax1.legend(wedges, legend_labels, loc="lower left", bbox_to_anchor=(-0.1, -0.12, 1, 1))
plt.tight_layout()
plt.show()
```

```python
# Calculate ride counts by user type and day
week_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
ride_counts_day = df.groupby(['day', 'member_casual'])['member_casual'].count().unstack().loc[week_order]

# Daily Ride Counts
fig, ax2 = plt.subplots(figsize=(12, 8))
ride_counts_day.plot(kind='bar', ax=ax2, edgecolor='black')
ax2.set_title('Daily Ride Counts', fontsize=18, fontweight='bold')
ax2.set_xlabel('', fontsize=14)
ax2.set_ylabel('Ride Count', fontsize=14)
ax2.legend(fontsize=12, loc='upper right')
ax2.grid(axis='y', linestyle='--', alpha=0.7)
ax2.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Calculate ride counts by user type and month
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
ride_counts_month = df.groupby(['month', 'member_casual'])['member_casual'].count().unstack().loc[month_order]

# Monthly Ride Counts
fig, ax3 = plt.subplots(figsize=(12, 8))
ride_counts_month.plot(kind='bar', ax=ax3, color=['#4c72b0', '#dd8452'], edgecolor='black')
ax3.set_title('Monthly Ride Counts', fontsize=18, fontweight='bold')
ax3.set_ylabel('Ride Count', fontsize=14)
ax3.set_xlabel('')
ax3.legend(fontsize=12)
ax3.grid(axis='y', linestyle='--', alpha=0.7)
ax3.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Ride Count by Hour
plt.figure(figsize=(12, 6))
sns.countplot(x='hour', hue='member_casual', data=df, edgecolor='black')
plt.title('Ride Count Across Hours', fontsize=16, fontweight='bold')
plt.xlabel('Hour', fontsize=14)
plt.ylabel('Ride Count', fontsize=14)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

#### **Insights**

- Since we cannot distinguish each user, it's unclear whether members ride more frequently or if there are simply more member users than casual ones.

**Daily Patterns:**        

- Members have higher ride counts on weekdays, while casual riders peak on weekends.

- Member riders consistently have higher ride counts across all days of the week. Peaks are observed on Tuesday and Thursday, indicating a regular usage pattern, possibly linked to work commutes.
        
- Casual riders have higher ride counts on Saturdays and Sundays, almost matching the counts of member riders on these days. This suggests that casual riders are more likely to use bikes for leisure or recreational activities during weekends.    

**Monthly Patterns:**

- Both casual users and members show increased ride counts from January, peaking in July and August, and then decreasing towards December. This trend highlights the seasonal popularity of bike usage, likely due to better weather conditions during the summer and fall.

**Hour Analysis:**

- Evening hours show a lot of traffic compared to other times, largely due to the office timings of member riders.

- Member usage peaks at 8 am (work start time) and 5 pm (end of workday).

- Casual bike usage increases slightly throughout the day, peaking at 5 pm and then decreasing.

```python
# Calculate mean ride duration by user type and day
mean_duration_day = df.groupby(['day', 'member_casual'])['duration_minutes'].mean().unstack().loc[week_order]

# Aggregation
agg = pd.DataFrame(df.groupby('member_casual')['duration_minutes']
                   .agg(['count', 'mean'])
                   .reset_index()
                   .rename(columns={'member_casual': 'user_type',
                                    'count': 'ride_count',
                                    'mean': 'mean_duration'}))

# Create the first two plots in a horizontal layout
fig, axes = plt.subplots(1, 2, figsize=(24, 12))

# Plot 1: Total Mean Ride Duration
axes[0].set_title('Mean Ride Duration (minutes) \n ', fontsize=36, fontweight='bold')
sns.barplot(data=agg, x='user_type', y='mean_duration', ax=axes[0], edgecolor='black')
axes[0].set_xlabel('', fontsize=36)
axes[0].set_ylabel('Minutes', fontsize=30, va='bottom')
axes[0].tick_params(axis='x', labelsize=30)
axes[0].yaxis.set_label_coords(-0.05, 0.15)
for index, row in agg.iterrows():
    axes[0].text(index, row['mean_duration'] + 0.2, f"{row['mean_duration']:.1f}",
                 color='black', ha="center", fontsize=30)

# Plot 2: Daily Mean Ride Duration
mean_duration_day.plot(kind='bar', ax=axes[1], edgecolor='black')
axes[1].set_title('Daily Mean Ride Duration \n ', fontsize=36, fontweight='bold')
axes[1].set_xlabel('', fontsize=80)
axes[1].set_ylabel('Minutes', fontsize=30, va='bottom')
axes[1].tick_params(axis='x', labelsize=20)
axes[1].yaxis.set_label_coords(-0.04, 0.10)
axes[1].legend(fontsize=26, loc='upper center')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
axes[1].tick_params(axis='x', rotation=0)
for p in axes[1].patches:
    axes[1].annotate(f"{p.get_height():.1f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='left', va='center', xytext=(0, 10), textcoords='offset points', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
# Calculate mean ride duration by user type and month
mean_duration_month = df.groupby(['month', 'member_casual'])['duration_minutes'].mean().unstack().loc[month_order]

# Plot 3: Monthly Mean Ride Duration
fig, ax = plt.subplots(figsize=(24, 12))
mean_duration_month.plot(kind='bar', ax=ax, color=['#4c72b0', '#dd8452'], edgecolor='black')
ax.set_title('Monthly Mean Ride Durations \n', fontsize=36, fontweight='bold')
ax.set_ylabel('Minutes', fontsize=30, va='bottom')
ax.yaxis.set_label_coords(-0.04, 0.10)
ax.set_xlabel('', fontsize=36)
ax.tick_params(axis='x', labelsize=30)
ax.legend(fontsize=26)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.tick_params(axis='x', rotation=45)
for p in ax.patches:
    ax.annotate(f"{p.get_height():.0f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=20)

plt.tight_layout()
plt.show()
```

<a id="monthly-chart"></a>
**Insights**

- On average, Casual users take rides almost twice as long as Members.

- Casual and member users‚Äô average ride duration peaks on Saturdays and Sundays.

- Casual riders' ride durations decrease by nearly half in December, likely due to colder temperatures, and gradually recover throughout January to peak in May. 

- In contrast, member riders' rides remain shorter and more consistent throughout the year, with a slight increase during the milder months.

<div style="border: 1px solid #e0e0e0; background-color: #fff3cd; padding: 20px; border-radius: 8px; margin-bottom: 15px;">
<h3 style="color:black;margin: 0;"><strong>Popular Routes</strong></h3>       
<a id="popular-routes"></a>

</div>

```python
# Filter rows where start_station_name or end_station_name are 'not_at_station'
df_station = df[(df['start_station_name'] != 'not_at_station') & (df['end_station_name'] != 'not_at_station')]

# Filter and group data by member type and routes
casual_routes = df_station[df_station['member_casual'] == 'casual'].groupby(['start_station_name', 'end_station_name']).agg({
    'ride_id': 'count',
    'distance_km': 'first'
}).reset_index()

member_routes = df_station[df_station['member_casual'] == 'member'].groupby(['start_station_name', 'end_station_name']).agg({
    'ride_id': 'count',
    'distance_km': 'first'
}).reset_index()

# Create route column
for routes_df in [casual_routes, member_routes]:
    routes_df['route'] = routes_df['start_station_name'] + ' --> ' + routes_df['end_station_name']

# Select top popular routes
top_casual_routes = casual_routes.sort_values(by='ride_id', ascending=False).head(15)
top_member_routes = member_routes.sort_values(by='ride_id', ascending=False).head(15)

# Plotting the most popular routes
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=False, figsize=(10,20))
sns.barplot(data=top_casual_routes, x='ride_id', y='route', ax=ax1, palette='Blues_r')
sns.barplot(data=top_member_routes, x='ride_id', y='route', ax=ax2, palette='YlOrBr_r')
ax1.set_title('Top 15 Popular Routes (Casual Riders) \n', fontsize=16, fontweight='bold')
ax1.set_xlabel('Number of Rides by Casual Riders', fontsize=14)
ax1.set_ylabel('Routes', fontsize=14)
ax2.set_title('Top 15 Popular Routes (Member Riders) \n', fontsize=16, fontweight='bold')
ax2.set_xlabel('Number of Rides by Member Riders', fontsize=14)
ax2.set_ylabel('Routes', fontsize=14)
plt.show()
```

**Insights**

- Within Casual riders, the top favored routes are all situated in the Loop, Chicago's central business district.

- Among Member riders, the top favored routes are around Ellis Ave. This area is home to the University of Chicago, indicating that university students frequently use bikes for campus travel.
